```python
import h5py
import os
#load hdf5 file
import pickle
import numpy as np
import matplotlib.pyplot as plt

file = "IC_0.csv"
file2 = "solution_0.csv"
file3 = "result.csv"
# load csv and show image
data = np.loadtxt(file, delimiter=',')[2:][::4]
data2 = np.loadtxt(file2, delimiter=',')[2:][::4]
result = np.loadtxt(file3, delimiter=',')[::4]
print(len(data), len(data2), len(result))
# now rearrange data as 128*128
data = data.reshape(128, 128)
data2 = data2.reshape(128, 128)
result = result.reshape(128, 128)
# flip all image upside down
data = np.flipud(data)
data2 = np.flipud(data2)
result = np.flipud(result)
plt.imshow(data, cmap='turbo', vmin = -1, vmax = 1,interpolation='nearest')
plt.title("Initial Condition")
plt.colorbar()
plt.show()
plt.imshow(data2, cmap='turbo', vmin = -1, vmax = 1, interpolation='nearest')
plt.title("Solution at T=1.0")
plt.colorbar()
plt.show()
plt.imshow(result,cmap='turbo', vmin = -1, vmax = 1, interpolation='nearest')
plt.title("Result from WebGL")
plt.colorbar()
plt.show()

```
