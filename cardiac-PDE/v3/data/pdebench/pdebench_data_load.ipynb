```python
import h5py
import os
#load hdf5 file
import pickle
import numpy as np
import matplotlib.pyplot as plt

file = "1D_Advection_Sols_beta0.1.hdf5"

with h5py.File(file, 'r') as f:
    
    # 2. See the names of the "folders" (datasets) inside
    print("Keys in file:", list(f.keys()))
    
    # 3. Access a specific dataset (replace 'tensor' with a key from step 2)
    data = f['tensor'][:]  # The [:] loads the data into a NumPy array
    tcoor = f['t-coordinate'][:]
    xcoor = f['x-coordinate'][:]
    print("Data shape:", data.shape)

# take random 100 samples from data
random_indices = np.random.choice(data.shape[0], size=100, replace=False)
random_samples = data[random_indices]
```

```python
plt.imshow(random_samples[0], aspect='auto')
```

```python
# download random_samples[0] at t=0 and t=-1 as csv file
np.savetxt("random_sample_t0.csv", np.tile(random_samples[0][0], (len(xcoor), 1)), delimiter=",")
np.savetxt("random_sample_t-1.csv", np.tile(random_samples[0][-1], (len(xcoor), 1)), delimiter=",")
```
