```python
from bs4 import BeautifulSoup

# 1. Your new shader code as a string
new_shader_code = """#version 300 es
precision highp float;
// ... your new logic here ...
void main() {
    ocolor = vec4(1.0, 0.0, 0.0, 1.0); 
}
"""

# 2. Load the HTML file
file_path = 'index.html'
with open(file_path, 'r', encoding='utf-8') as f:
    soup = BeautifulSoup(f, 'html.parser')

# 3. Find the script tag by ID and replace its string content
target_script = soup.find('script', id='march')
if target_script:
    target_script.string = new_shader_code
    
    # 4. Save the modified file
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(str(soup))
    print("Shader replaced successfully!")
else:
    print("Could not find a script tag with id='march'")
```
