```python
import numpy as np
import os
def transform_csv_to_ndarrays(path, N=512):
    # Load the flat CSV data
    # Assuming the file is a single row or a list of values separated by commas
    data = np.genfromtxt(path, delimiter=',')
    
    # Each point x has 4 values: u, v, w, and 0
    # Reshape the flat array into (N*N, 4)
    reshaped_data = data.reshape(-1, 4)
    
    # Extract U, V, and W components (columns 0, 1, and 2)
    U_flat = reshaped_data[:, 0]
    V_flat = reshaped_data[:, 1]
    W_flat = reshaped_data[:, 2]
    
    # Reshape into (1, N, N)
    U = U_flat.reshape(1, N, N)
    V = V_flat.reshape(1, N, N)
    W = W_flat.reshape(1, N, N)
    
    return U, V, W

folder = './tau_d_0.5714'

# count how many csv files in folder
csv_count = len([f for f in os.listdir(folder) if f.endswith('.csv')])

U_batch = np.zeros((csv_count, 512, 512))
V_batch = np.zeros((csv_count, 512, 512))
W_batch = np.zeros((csv_count, 512, 512))

for i, f in enumerate(os.listdir(folder)):
    if f.endswith('.csv'):
        path = os.path.join(folder, f)
        U, V, W = transform_csv_to_ndarrays(path)
        U_batch[i] = U
        V_batch[i] = V
        W_batch[i] = W
print(U_batch.shape, V_batch.shape, W_batch.shape)
np.savez_compressed('./tau_d_0.5714/UVW_array_data.npz', U=U_batch, V=V_batch, W=W_batch)
```
